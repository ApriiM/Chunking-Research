# 🔍 Chunker Qualitative Analysis
Narzędzie do porównywania jakości różnych chunkerów na poziomie konkretnych pytań i fragmentów dokumentów.

## 0. Importy i ścieżki

```python
import sys
sys.path.insert(0, "..")  # dostosuj jeśli moduł jest w innym miejscu

import warnings
warnings.filterwarnings("ignore")

import pandas as pd
pd.set_option("display.max_colwidth", 200)
pd.set_option("display.max_rows", 100)

from chunker_analysis import run_analysis, print_groups, filter_groups, build_dataframe
```

---
## 1. ⚙️ Konfiguracja

```python
# ── Ścieżki ───────────────────────────────────────────────
PIRB_DATA_DIR    = "./data/pirb/pirb_data"
RETRIEVED_DIR    = "./data/pirb/retrieved_documents"
RESULTS_DIR      = "./data/pirb/results"         # opcjonalnie do metryk
PROCESSED_DIR    = "./data/processed"

# ── Filtry ────────────────────────────────────────────────
TARGET_DATASETS         = ["gutenqa_all", "gutenqa_all_merged"]   # None = wszystkie
EXCLUDE_CHUNKERS        = []                # np. ["fixed_size", "sentence"]

# Pokazuj grupę tylko gdy >=1 chunker ma retrieved_relevant==True
ONLY_WHEN_RETRIEVED_RELEVANT = True

# Ile znaków z początku/końca relevantnych chunków musi być identycznych
# 0 = wyłączony filtr
MIN_COMMON_PREFIX = 3    # np. 5
MIN_COMMON_SUFFIX = 0   # np. 0

# ── Wyświetlanie ──────────────────────────────────────────
MAX_CHUNK_CHARS = 400    # maks. znaków tekstu chunka w princie
SHOW_SCORES     = True
MAX_GROUPS      = None   # None = wszystkie; ustaw np. 20 do podglądu
```

---
## 2. 🚀 Uruchomienie analizy

```python
groups, chunk_lookup, df = run_analysis(
    pirb_data_dir             = PIRB_DATA_DIR,
    retrieved_dir             = RETRIEVED_DIR,
    processed_dir             = PROCESSED_DIR,
    target_datasets           = TARGET_DATASETS,
    exclude_chunkers          = EXCLUDE_CHUNKERS,
    only_when_retrieved_relevant = ONLY_WHEN_RETRIEVED_RELEVANT,
    min_common_prefix         = MIN_COMMON_PREFIX,
    min_common_suffix         = MIN_COMMON_SUFFIX,
    max_chunk_chars           = MAX_CHUNK_CHARS,
    show_scores               = SHOW_SCORES,
    max_groups                = MAX_GROUPS,
    print_results             = True,
)
```

```python
print_groups(groups, chunk_lookup, MAX_CHUNK_CHARS, SHOW_SCORES, MAX_GROUPS)
```

---
## 3. 📊 DataFrame – interaktywne przeglądanie

```python
print(f"Kolumny: {list(df.columns)}")
print(f"Kształt: {df.shape}")
df.head(10)
```

### 3a. Filtruj tylko zwrócone relevantne chunki

```python
df_retrieved = df[df["is_retrieved"] == True].copy()
df_retrieved.sort_values(["query_id", "chunker_name", "retrieved_rank"], inplace=True)
df_retrieved[["query_id", "chunker_name", "chunk_id", "retrieved_rank", "retrieved_score", "chunk_text"]].head(20)
```

### 3b. Chunki, które NIE zostały zwrócone (miss)

```python
df_missed = df[df["is_retrieved"] == False].copy()
print(f"Przeoczone relevantne chunki: {len(df_missed)}")
df_missed[["query_id", "question", "chunker_name", "chunk_id", "chunk_text"]].head(20)
```

### 3c. Porównanie chunkerów – hit rate (czy relevantny chunk znalazł się w retrieved)

```python
hit_rate = (
    df.groupby("chunker_name")["is_retrieved"]
    .agg(total="count", hits="sum")
    .assign(hit_rate=lambda x: (x["hits"] / x["total"] * 100).round(2))
    .sort_values("hit_rate", ascending=False)
)
print("Hit rate relevantnych chunków per chunker:")
hit_rate
```

### 3d. Średnia pozycja (rank) zwróconych relevantnych chunków per chunker

```python
rank_stats = (
    df[df["is_retrieved"] == True]
    .groupby("chunker_name")["retrieved_rank"]
    .agg(["mean", "median", "min", "max", "count"])
    .round(2)
    .sort_values("mean")
)
print("Statystyki ranku dla zwróconych relevantnych chunków:")
rank_stats
```

---
## 4. 🔎 Analiza konkretnego pytania

```python
# Wyszukaj grupy zawierające fragment pytania
SEARCH_PHRASE = ""  # wpisz fragment pytania

if SEARCH_PHRASE:
    matching = [
        g for g in groups
        if SEARCH_PHRASE.lower() in g.question.lower()
    ]
    print(f"Znaleziono {len(matching)} grup z frazą '{SEARCH_PHRASE}'")
    print_groups(matching, chunk_lookup, MAX_CHUNK_CHARS, SHOW_SCORES)
else:
    print("Ustaw SEARCH_PHRASE żeby wyszukać konkretne pytanie.")
```
